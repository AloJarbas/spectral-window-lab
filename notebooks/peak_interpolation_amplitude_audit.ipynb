# Peak interpolation amplitude audit

This notebook is the slower companion to the peak-interpolation sidecar. The question is no longer whether flat-top has a coarse sampled-peak trap. The question is whether one tiny local interpolator changes the amplitude ranking enough to open a compact alternative.


## Local estimators

Given three neighboring magnitude samples `y_{-1}, y_0, y_{+1}`, a quadratic peak fit uses

$$p = \frac{1}{2}\frac{y_{-1} - y_{+1}}{y_{-1} - 2y_0 + y_{+1}}, \qquad \hat y = y_0 - \frac{1}{4}(y_{-1} - y_{+1})p.$$

This study tries that formula in two lanes: once on the magnitudes themselves and once on `\log |X[k]|`, then exponentiates back. The difference turns out to matter.


```python
from windowlab.peak_interpolation import study_peak_interpolation

study = study_peak_interpolation(length=129, probe_ffts=(256, 512, 1024, 2048, 4096), highlight_fft=256)
[(row.window, row.estimator, row.fft_size, round(row.worst_abs_bias_db, 6)) for row in study.rows if row.fft_size == 256]
```

## Figure

![Peak interpolation audit](../art/window-peak-interpolation-audit.png)

Read the Blackman-Harris panel first. The raw sampled peak leaves a visible amplitude hole. The linear parabola helps. The log parabola nearly wipes the hole out.


```python
rows_256 = [row for row in study.rows if row.fft_size == 256]
[(row.window, row.estimator, round(row.worst_abs_bias_db, 6), round(row.enbw_bins, 3)) for row in rows_256]
```

## Why this changes the map

The raw sampled-peak lane still belongs to flat-top. But if one tiny 3-point log fit is allowed, Blackman-Harris becomes a serious compact amplitude choice: much lower ENBW than flat-top, yet already down in the few-millith-decibel bias range in this bounded audit.


## Caveats

1. This is noiseless and isolated-tone only.
2. It compares one local interpolator family, not every estimator.
3. The conclusion is conditional: flat-top still wins when you want the sampled peak itself to be honest with no extra work.
