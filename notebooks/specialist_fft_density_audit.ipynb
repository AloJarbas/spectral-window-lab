```python
from windowlab.specialist_density import study_specialist_fft_density

study = study_specialist_fft_density(length=129, reference_fft=16384, probe_ffts=(256, 512, 1024, 2048, 4096))
[(row.window, row.fft_size, round(row.peak_sidelobe_error_db, 3), round(row.main_lobe_width_error_bins, 3)) for row in study.rows if row.fft_size == 512]
```

## The bounded result

That changes the repo in a useful way because it kills the lazy conclusion that “deep sidelobe” is one numerical-measurement class.
